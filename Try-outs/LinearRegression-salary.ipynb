```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("https://data.alexmattyou.repl.co/db/salaryData2d.csv")#dataset coppied from Kaggle
df_x,df_y=data_set[['YearsExperience']],data_set['Salary']
plt.scatter(df_x,df_y)
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.grid()
plt.xlim(0,12)
plt.ylim(30000,130000)
plt.show()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
reg=linear_model.LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(df_x,df_y,test_size=0.2,random_state=4)
reg.fit(x_train.values,y_train.values)
```

```python
a = np.arange(0,15).reshape(-1,1)
b = reg.predict(a)
plt.plot(df_x,df_y,'o')
plt.plot(a,b)
plt.xlabel("years of experience")
plt.ylabel("salary")
plt.grid()
plt.xlim(0,12)
plt.ylim(30000,130000)
plt.show()
```
