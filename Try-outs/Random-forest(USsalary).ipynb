```python
import numpy as np
import pandas as pd

train = pd.read_csv('https://data.alexmattyou.repl.co/db/US-salary/train.csv')
test = pd.read_csv('https://data.alexmattyou.repl.co/db/US-salary/test.csv')

train.info()
```

```python
print ("The train data has",train.shape)
print ("The test data has",test.shape)
```

```python
train.head()
```

```python
train.isnull().sum()
```

```python
train.workclass.fillna('Private',inplace=True)
train.occupation.fillna('Prof-specialty',inplace=True)
train['native.country'].fillna('United-States',inplace=True)
```

```python
(train.target.value_counts()/train.shape[0])*100
```

```python
(pd.crosstab(train.education, train.target,margins=True)/train.shape[0])*100
```

```python
from sklearn import preprocessing

for x in train.columns:
    if train[x].dtype == 'object':
        lbl = preprocessing.LabelEncoder()
        lbl.fit(list(train[x].values))
        train[x] = lbl.transform(list(train[x].values))
```

```python
train.head(100)
```

```python
y = train['target']
del train['target']
x = train
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=500,max_depth=6)
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
prediction = clf.predict(x_test)
acc =  accuracy_score(np.array(y_test),prediction)
print('accuracy:',acc*100)
```
