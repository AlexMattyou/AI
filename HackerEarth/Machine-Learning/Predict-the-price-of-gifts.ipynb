```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#importing basic modules
```

```python
#link given train and test data
train = pd.read_csv('https://data.alexmattyou.repl.co/db/Predict-the-price-of-gifts/train.csv')
test = pd.read_csv('https://data.alexmattyou.repl.co/db/Predict-the-price-of-gifts/test.csv')
print("The train data has",train.shape)
print(train.info())
```

```python
train.isnull().sum()
# wow no null values
```

```python
train['instock_year'] = pd.DatetimeIndex(train['instock_date']).year
train['instock_month'] = pd.DatetimeIndex(train['instock_date']).month
train['stock_update_year'] = pd.DatetimeIndex(train['stock_update_date']).year
train['stock_update_month'] = pd.DatetimeIndex(train['stock_update_date']).month
train['uk_'] = pd.DatetimeIndex(train['uk_date1']).year
train['stock_update_month'] = pd.DatetimeIndex(train['uk_date1']).month
train.drop(columns=['gift_id','instock_date','stock_update_date'],inplace=True)
```

```python
# Time to train the model
y = train['High_Cap_Price']
del train['High_Cap_Price']
x = train
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```
