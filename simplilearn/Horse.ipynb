```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv('https://data.alexmattyou.repl.co/db/simpillearn-demo/Lesson-6/horse.csv')
db
```

```python
db.isnull().sum()
```

```python
for i in db.columns:
  if db[i].isnull().sum() > 0:
    print(db[i].value_counts().head(5))
```

```python
from sklearn.preprocessing import LabelEncoder
for i in db.columns:
  if db[i].dtype == 'object':
    le = LabelEncoder()
    le.fit(list(db[i].values))
    db[i] = le.transform(list(db[i].values))
```

```python
from sklearn.impute import SimpleImputer
db = db.fillna(0.01010101)
imp = SimpleImputer(missing_values=0.01010101,strategy='most_frequent')
imp.fit(db)
```

```python
db.isnull().sum()
```

```python
y = db['outcome']
del db['outcome']
X = db
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,train_size=.7,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
alg = DecisionTreeClassifier()
alg.fit(Xtrain, ytrain)
ypred = alg.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
# so poor
```

```python
from sklearn.ensemble import RandomForestClassifier
alg2 = RandomForestClassifier()
alg2.fit(Xtrain, ytrain)
ypred = alg2.predict(Xtest)
accuracy_score(ypred,ytest)
```

found RandomForestClassifier works better than DecisionTreeClassifier for this data
