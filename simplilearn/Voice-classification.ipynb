```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
db = pd.read_csv('https://data.alexmattyou.repl.co/db/simpillearn-demo/Lesson-6/voice-classification.csv')
db
```

```python
db.isnull().sum()
```

```python
y = db['label']
del db['label']
X = db
```

```python
from sklearn.preprocessing import LabelEncoder
for i in db.columns:
  if db[i].dtype == 'object':
    le = LabelEncoder()
    le.fit(list(db[i].values))
    db[i] = le.transform(list(db[i].values))
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=0)
```

```python
from sklearn.svm import SVC
alg = SVC()
mod = alg.fit(X_train, y_train)
```

```python
y_pred = alg.predict(X_test)
from sklearn import metrics
import seaborn as sns
cf_matrix = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(cf_matrix, annot=True)
```
